```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
url1 = 'https://raw.githubusercontent.com/mandy941/Chinese-Product-in-US-Market/main/SNACK%20MARKET/Analysis%20Data/candy/Yami%20Candy.csv'
url2 = 'https://raw.githubusercontent.com/mandy941/Chinese-Product-in-US-Market/main/SNACK%20MARKET/Analysis%20Data/candy/Candy-Weee!-Snacks-ScrapingData-ScrapeStorm.csv'
url3 = 'https://raw.githubusercontent.com/mandy941/Chinese-Product-in-US-Market/main/SNACK%20MARKET/Analysis%20Data/candy/candy-freshgogoData-ScrapeStorm.csv'
yami = pd.read_csv(url1, encoding='latin1')
yami.head()
```

```python
weee = pd.read_csv(url2, encoding='latin1')
weee.head()
```

```python
freshgogo = pd.read_csv(url3)
freshgogo.head()
```

# Combine three tables together

```python
candy = pd.concat([yami,freshgogo,weee])
candy.head()
```

```python
# Finding the correlation bewteen the Features.
plt.figure(figsize=(8,5))

corr = candy.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, 
            annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
plt.title('Figure 5: Correlations')
```

## Origin vs Platform

```python
# Origin vs Platform

plt.rcParams['figure.figsize'] = (10, 6)

# compare the year with Origin with Platform
plt.subplot(121)
sns.countplot(x='Platform',hue= 'Origin', data=candy, palette = 'husl')
plt.legend(bbox_to_anchor=(-0.2, 0.5), loc='right', borderaxespad=0)
plt.title('Platform vs. Origin')
```

```python
# check the Origin percentage of Platform
plt.rcParams['figure.figsize'] = (10, 6)
ax= plt.subplot(122)
c2_c = pd.crosstab(candy.Platform, candy.Origin)
c2_c.div(c2_c .sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, ax=ax)
plt.title('Origin percentage by Platform')
plt.legend(bbox_to_anchor=(-0.2, 1), loc='left', borderaxespad=0)
plt.xlabel('Platform')
plt.xticks(rotation=0)
plt.ylabel('Percentage of Origin')

c2_c
```

## Origin Vs Sold_Weekly_Group

```python
candy['Sold_Weekly_Group'] = pd.cut(x = candy['Sold_Weekly'], 
                                   bins = [0,50,100,200,300,400,500,1000,2000,10000],
                                   labels = ['Less than 50','50+','100+','200+','300+','400+','500+','1000+','2000+'])
candy.head()
```

```python
# origin vs sold weekly group

plt.rcParams['figure.figsize'] = (20, 6)

# compare the year with Origin with Platform
plt.subplot(121)
sns.countplot(x='Sold_Weekly_Group',hue= 'Origin', data=candy, palette = 'husl')
plt.legend(bbox_to_anchor=(-0.2, 0.5), loc='right', borderaxespad=0)
plt.title('Sold_Weekly_Group vs. Origin')
```

```python
# origin vs sold weekly group

plt.rcParams['figure.figsize'] = (20, 7)
# check the Origin percentage of Platform
ax= plt.subplot(122)
c2_c = pd.crosstab(candy.Sold_Weekly_Group, candy.Origin)
c2_c.div(c2_c .sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, ax=ax)
plt.title('Origin percentage by Sold_Weekly_Group')
plt.legend(bbox_to_anchor=(-0.2, 1), loc='left', borderaxespad=0)
plt.xlabel('Sold_Weekly_Group')
plt.xticks(rotation=0)
plt.ylabel('Percentage of Origin')

c2_c
```

```python
# origin vs sold weekly group

plt.rcParams['figure.figsize'] = (20, 6)

# compare the year with Origin with Platform
plt.subplot(121)
sns.countplot(x='Price_Group',hue= 'Origin', data=candy, palette = 'husl')
plt.legend(bbox_to_anchor=(-0.2, 0.5), loc='right', borderaxespad=0)
plt.title('Price_Group vs. Origin')
```

```python
# origin vs sold weekly group

plt.rcParams['figure.figsize'] = (20, 7)

# check the Origin percentage of Platform
ax= plt.subplot(122)
c2_c = pd.crosstab(candy.Price_Group, candy.Origin)
c2_c.div(c2_c .sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, ax=ax)
plt.title('Origin percentage by Price_Group')
plt.legend(bbox_to_anchor=(-0.2, 1), loc='left', borderaxespad=0)
plt.xlabel('Price_Group')
plt.xticks(rotation=0)
plt.ylabel('Percentage of Origin')

c2_c
```
